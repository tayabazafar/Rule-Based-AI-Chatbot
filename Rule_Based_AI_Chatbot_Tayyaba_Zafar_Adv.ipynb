```python
#Starting chatbot
print("=" * 50)
print("AI Rule-Based Chatbot")
print("Welcome! I'm Decodebot.")
print('Type "help" to see all available commands.')
print("=" * 50)
#Ask the user for their name
name = input("Bot: What's your name? ")
print(f"Bot: Nice to meet you, {name}!")
#Keep the chatbot running until the user exits
while True:
  message = input("You: ").lower() #Get user message in lower case
  #Respond to the greetings
  if message in ["hi", "hello", "hey"]:
   print(f"Bot: Hello {name}! How are you?")
  elif message == "fine" or message == "good":
    print("Bot: That's great to hear!")
  elif message == "how are you":
    print("Bot: I'm doing great! Thanks for asking.")
  elif message == "who are you":
    print("Bot: I'm DecodeBot, a rule-based AI chatbot developed in Python.")
  #Show available chatbot commands
  elif message == "help":
    print("=" * 50)
    print("📖 HELP MENU")
    print("=" * 50)
    print("Bot: Available commands:")
    print("- hi, hey or hello")
    print("- fine or good")
    print("- thank you")
    print("- who created you")
    print("- good morning")
    print("- good night")
    print("- bye")
    print("- what can you do")
    print("- how are you")
    print("- who are you")
    print("- motivate me")
    print("- tell me a joke")
    print("- tell me another joke")
    print("- python")
    print("- github")
  #Respond to the user request
  elif message == "what can you do":
    print("Bot: I can respond to greetings, thank you messages, and a few predefined questions. Type 'help' to see all available commands.")
  elif message == "motivate me":
    print("Bot: Every expert was once a beginner. Keep learning and never give up!")
  elif message == "tell me a joke":
    print("Bot: Why do programmers prefer dark mode? Because light attracts bugs! 😂")
  elif message == "tell me another joke":
    print("Bot: There are only 10 kinds of people: those who understand binary and those who don't. 😂")
  elif message in ["python", "what is python"]:
    print("Bot: Python is a beginner-friendly programming language used in AI, web development, and automation.")
  elif message in ["github", "what is github"]:
    print("Bot: GitHub is used to store and showcase your programming projects.")
  elif message in ["thank you", "thanks", "thx"]:
    print(f"Bot: You are welcome {name}.")
  elif message == "who created you":
    print("Bot: I was created by Tayyaba.")
  elif message == "good morning":
    print(f"Bot: Good morning {name}! Have a nice day.")
  elif message == "good night":
    print(f"Bot: Good night {name}! Sweet Dreams.")
  #Exit the chatbot
  elif message in ["bye", "exit", "quit"]:
    print(f"Bot: Goodbye {name}! Have a nice day.")
    break
  #Handle the unknown user input
  else :
    print(f"Bot: Sorry {name}, I don't recognize that command. Type 'help' to see what I can do.")
```
